```python
#z_teat using p_value

import pandas as pd
import numpy as np
import scipy.stats as stats

# Load the dataset
df = pd.read_csv("California Wildfire Damage.csv")

# Extract the column
data = df["Businesses_Destroyed"]

# Given or calculated values
population_mean = data.mean()  # Assuming the population mean is the dataset's mean
population_std = data.std()    # Assuming this is the population standard deviation
sample_size = len(data)        # Sample size

# Hypothetical sample mean
sample_data = df[df["Location"] == "Los Angeles"]["Businesses_Destroyed"]
sample_mean=sample_data.mean()

# Compute Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Find p-value
p_value = 2 * (1 - stats.norm.cdf(np.abs(z_score)))  # Two-tailed test

# Significance level
alpha = 0.05

# Print results
print(population_mean)
print(population_std)
print(sample_size)
print(sample_mean)
print("Z-score:", z_score)
print("P-value:", p_value)

# Decision
if p_value < alpha:
    print("Reject the null hypothesis (H0) - Significant difference found.")
else:
    print("Fail to reject the null hypothesis (H0) - No significant difference.")
```

```python
#Z_Test using Z_ Critical

import pandas as pd
import numpy as np
import scipy.stats as stats


#Load dataset using pandas
df=pd.read_csv("California Wildfire Damage.csv")

data=df["Businesses_Destroyed"]
population_mean=data.mean()
population_std=data.std()
sample_size=len(data)
sample_data = df[df["Location"] == "Shasta"]["Businesses_Destroyed"]
sample_mean=sample_data.mean()
alpha=0.05

#z_score
z_score=(sample_mean-population_mean)/(population_std/np.sqrt(sample_size))

#z_critical
z_critical=stats.norm.ppf(1-alpha/2)


print(population_mean)
print(population_std)
print(sample_size)
print(sample_mean)
print("Z-score:",z_score)
print("z_critical:",z_critical)

if z_score>z_critical:
  print("Reject the null hypothesis (H0) - Significant difference found.")
else:
  print("Fail to reject the null hypothesis (H0) - No significant difference.")
```

```python
#T_TEST USING P_VALUE
import pandas as pd
import numpy as np
import scipy.stats as stats

# Loading  dataset
df = pd.read_csv("HR-Employee-Attrition.csv")

# Splitting the dataset into two groups based on EducationField
group_life_sciences = df[df["EducationField"] == "Life Sciences"]["Education"]
group_other = df[df["EducationField"] != "Life Sciences"]["Education"]

# Performing Independent T-Test
t_stat, p_value = stats.ttest_ind(group_life_sciences, group_other, nan_policy='omit')

# Display results
print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("Significant difference found between Life Sciences and other education fields or Rejects the null hypothesis.")
else:
    print("No significant difference found between Life Sciences and other education fields  or Fail to rejects the null hypothesis")

#Performing One Sample T-Test
sample_data=df[df["EducationField"]=="Life Sciences"]["Education"]
population_mean=df["Education"].mean()

t_stat, p_value= stats.ttest_1samp(sample_data, popmean=population_mean, nan_policy="omit")

print(population_mean)
print(f"t-statistic:{t_stat:.4f},p-value:{p_value:.4f} ")

if p_value<0.05:
  print("Significant difference found between Life Sciences and other education fields or Rejects the null hypothesis.")
else:
  print("No significant difference found between Life Sciences and other education fields  or Fail to rejects the null hypothesis")
```

```python
#Chi-Squared Test
import pandas as pd
import numpy as np
import scipy.stats as stats

# Load dataset
df = pd.read_csv("HR-Employee-Attrition.csv")

# Create a contingency table (e.g., Department vs. DailyRate)
contingency_table = pd.crosstab(df["Department"], df["DailyRate"])

# Perform Chi-Squared Test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

# Display results
print(f"Chi-Squared Statistic: {chi2_stat:.4f}, P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies Table:\n", expected)

# Interpretation
if p_value < 0.05:
    print("There is a significant relationship between EducationField and Attrition.")
else:
    print("No significant relationship found between EducationField and Attrition.")
```

```python
#ANOVA or F-TEST
import pandas as pd
import scipy.stats as stats

# Load dataset
df = pd.read_csv("HR-Employee-Attrition.csv")

# Selecting numerical dependent variable (Example: MonthlyIncome)
# Grouping by categorical independent variable (Example: EducationField)
group1 = df[df["EducationField"] == "Life Sciences"]["MonthlyIncome"]
group2 = df[df["EducationField"] == "Medical"]["MonthlyIncome"]
group3 = df[df["EducationField"] == "Marketing"]["MonthlyIncome"]

# Performing One-Way ANOVA
f_stat, p_value = stats.f_oneway(group1, group2, group3)

# Display Results
print(f"F-statistic: {f_stat:.4f}, P-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("Significant difference found between the groups.")
else:
    print("No significant difference found between the groups.")
```
